```python
%matplotlib inline

import csv
import os
import zipfile
from io import StringIO
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing

le = preprocessing.LabelEncoder()


data_path = "./"
full_path = data_path+"/train.csv.zip"
base = os.path.basename(data_path+"train.csv.zip")
file_name = os.path.splitext(base)[0]
with open(full_path, 'rb') as fh:
    zfile = zipfile.ZipFile(fh)
    data = pd.read_csv(zfile.open(file_name))
le.fit(data.iloc[:,-1])
print(data.describe())
```

```python
for i in range(1,data.shape[0]-3):
    print(i)
    plt.scatter(data.iloc[:,0],data.iloc[:,i],c=le.transform(data.iloc[:,-1]))
    plt.show()
```

```python
import sklearn.ensemble
from sklearn.cross_validation import cross_val_score
import hyperopt

def run_cls(settings):
    print("settings")
    print("")
    clsif = settings.pop('type')
    clsif_opts = settings['options']
    for kys, it in zip(clsif_opts.keys(),clsif_opts.values()):
        if (it.is_integer()):
            settings['options'][kys] = int(it)
    
    rfc = clsif(**clsif_opts)
    c_val = -1 * cross_val_score(rfc,data.iloc[:,1:-1],le.transform(data.iloc[:,-1]),scoring='log_loss')
    return {
        'loss': np.mean(c_val),
        'status': hyperopt.STATUS_OK
        }

trials = hyperopt.Trials()
hyperopt.fmin(run_cls,
              space=hyperopt.hp.choice('Classifier',
                                        [
                                        {'type': sklearn.ensemble.RandomForestClassifier,
                                         'options': {
                                            'n_estimators': hyperopt.hp.quniform('rf_n_est', 10,15, 1),
                                            'n_jobs': 4.0}
                                        },
                                        {'type': sklearn.ensemble.AdaBoostClassifier,
                                         'options': {'n_estimators': hyperopt.hp.quniform('ada_n_est', 10, 20, 10)}},
                                        {'type': sklearn.ensemble.GradientBoostingClassifier,
                                         'options': {'n_estimators': hyperopt.hp.quniform('gd_n_est', 10, 200, 10)}},
                                        ]),
              algo = hyperopt.tpe.suggest,
              max_evals=5,
              trials = trials)
```

```python
import pandas as pd
print(trials.best_trial['result']['loss'])
print(zip(pd.DataFrame(trials.miscs)['vals'].values,trials.results))
```
